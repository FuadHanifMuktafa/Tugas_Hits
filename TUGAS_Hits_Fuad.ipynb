```python
import networkx as nx 
import matplotlib.pyplot as plt 
def main():
    
    G = nx.read_edgelist('uudasarhukum-csv.txt', create_using=nx.DiGraph())
    deg = dict(G.in_degree()) #Returns a dictionay, with key as nodes and indegrees as values.
    h,a = nx.hits(G)
    for i in deg.keys():
            print(h,a)
main()
```
